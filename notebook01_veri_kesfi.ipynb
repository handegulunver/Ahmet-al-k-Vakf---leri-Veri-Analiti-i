```python
file_path = "../data/enerji_verisi.xlsx.xlsx"

xls = pd.ExcelFile(file_path)
print("Sheet isimleri:")
print(xls.sheet_names)
```

```python
df_tahsilat = pd.read_excel(file_path, sheet_name="Tahsilat")
df_tahsilat_1 = pd.read_excel(file_path, sheet_name="Tahsilat 1")

df_tahakkuk = pd.read_excel(file_path, sheet_name="Tahakkuk")
df_tahakkuk_1 = pd.read_excel(file_path, sheet_name="Tahakkuk 1")
df_tahakkuk_2 = pd.read_excel(file_path, sheet_name="Tahakkuk 2")
```
